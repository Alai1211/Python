### 将日期设置为索引

```python
wind['Yr_Mo_Dy']=pd.to_datetime(wind['Yr_Mo_Dy'])
```

```python
wind=wind.set_index('Yr_Mo_Dy',drop = True)
```

```python
month = lambda date:date.month

wind['month']=wind['date'].apply(month)
```
