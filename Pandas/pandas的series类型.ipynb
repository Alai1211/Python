# panda库的series类型

将数据和索引一起显示的类型

带标签数组

基本操作类似字典


## 如何创建

python列表，index与列表元素个数一致

从标量值创建，index表达Series类型的尺寸

python字典，键值对中的‘键’是索引，index从字典中进行选择操作

ndarray，索引和数据都可以通过ndarray类型创建

其他函数，range()函数等

### 从列表进行创建

直接给定值即可[]

```python
import pandas as pd
```

```python
b = pd.Series([4,5,6,7])
```

```python
b
```

### 从标量值创建

需要对结构进行指定

index参数不能省略

```python
#创建一组值为25的series类型
s = pd.Series(25,index = ['a','b','c'])
```

```python
s
```

### 从字典创建

直接将字典赋给series

可通过index重新设定标签

```python
e = pd.Series({'a':9,'b':8,"c":7})
#index从字典中挑玄相关值创建
```

```python
e
```

```python
#修改标签
e = pd.Series({'a':9,'b':8,"c":7},index=['c','a','b','d'])
```

```python
e
```

### 从ndarray类型创建

同样index的值也可以通过ndarray数组指定

```python
import numpy as np
```

```python
n = pd.Series(np.arange(5))
```

```python
n
```

```python
n = pd.Series(np.arange(5),index=np.arange(9,4,-1))
#生成9到5，倒叙的suoyin
```

```python
n
```

### series的索引

索引方式,采用[] 

Series类型始终是索引加值的一种方式

series类型自己存在默认的数字索引（在所创建的索引不为数字的情况下）使用字符串索引的时候要带引号

自动索引和自定义索引不能并存


```python
e
```

```python
e[1]
```

```python
e['a']
```

```python
e[['c','a']]
```

```python
e[['c','a',0]]
#自动索引和自定义索引不能并存
```

```python
#自动索引切片
#索引为数字时，只有切片能使用自动索引的数字
n[:3]
```

```python
#任何计算都返回完整的索引+值的形式
np.exp(b)
```


### 【对齐】series类型的计算

索引相同的进行运算

series类型会自动对其不同索引的数据

```python
a = pd.Series([1,2,3],index = ['c','a','e'])
b = pd.Series([9,8,7,5],['a','b','c','d'])
```

```python
a+b
```
