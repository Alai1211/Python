```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

MovieLens 1M数据集含有来自6000名用户对4000部电影的100万条评分数据

包含用户信息、评分信息、电影信息

```python
unames=['user_id','gender','age','occupation','zip']
users=pd.read_csv('users.dat',sep='::',names=unames)
```

```python
users
```

```python
rnames=['user_id','movie_id','rating','timestamp']
rating=pd.read_csv('ratings.dat',sep='::',names=rnames)
```

```python
rating
```

```python
mnames=['movie_id','title','genres']
movies=pd.read_csv('movies.dat',sep='::',names=mnames,encoding='ISO-8859-1')
#此处出现'utf-8' codec can't decode byte 报错，加一个参数encoding改成如上
```

```python
movies.head()
```

表连接

连接三个表

```python
pd.merge(users,rating)
```

```python
data=pd.merge(pd.merge(rating,users),movies)
```

```python
data.head()
```

```python
#查看列信息
data.info()
```

### 不同性别对电影评价是否有差异

设置索引为电影名称，观察男性女性对不同电影的评价的均值

```python
mean_rating=pd.pivot_table(data,index='title',columns='gender',values='rating',aggfunc='mean')
#aggfunc默认求均值
```

```python
mean_rating[:5]
```

过滤掉评分不足250条的电影数据

```python
rating_by_title=data.groupby('title').size()
#Series类型，此处title就是索引
```

```python
rating_by_title.head()
```

```python
active_titles=rating_by_title.index[rating_by_title>=250]
```

```python
#获取评论数大于250的电影名称
active_titles
```

```python
#从mean rating中选取所需行
mean_rating=mean_rating.loc[active_titles]
```

```python
mean_rating
```

```python
#在女性中平均评价最高的电影
top_female=mean_rating.sort_values('F',ascending=False)
```

```python
top_female.head()
```

### 男性女性分歧最大的电影是哪部

利用平均得分之差

```python
mean_rating['diff']=mean_rating['M']-mean_rating['F']
```

```python
mean_rating.head()
```

```python
sort_diff=mean_rating.sort_values('diff')
```

```python
#负数说明女生更喜欢，正数说明男生更喜欢
sort_diff
```

```python
#女生更喜欢前十
sort_diff[:10]
```

```python
#男生更喜欢前十
sort_diff[::-1][:10]
```

### 无关性别的拥有最大争议的电影

```python
#计算评价的方差
rating_std_by_title = data.groupby('title')['rating'].std()
```

```python
#挑出其中男女争议最大的几部片子
rating_std_by_title = rating_std_by_title.loc[active_titles]
```

```python
#降序排列
rating_std_by_title.sort_values(ascending=False).head()
```
